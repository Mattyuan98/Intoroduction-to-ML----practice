### 1. Stacked Bar Chart<a name="1"></a>

- creating a stacked bar plot with error bars using bar.Note the parameters yerr used for error bars, and bottom to stack the women's bars on top of the men's bars.

````python
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
````

````python
labels = ['G1','G2','G3','G4','G5']
men_means = [20,35,30,35,27]
women_means = [25,32,34,20,25]
men_std = [2,3,4,1,2]
women_std = [3,5,2,3,3]
width = 0.35 #the width of the bars, can also be len(x) sequence
````

````python
fig,ax = plt.subplots()

ax.bar(labels, men_means,width,yerr = men_std, label = 'Men')
ax.bar(labels, women_means,width,yerr = women_std,bottom = men_means, label = 'Women')

ax.set_ylabel('Scores')
ax.set_title('Scores by group and gender')
ax.legend()

plt.show()
````

### 2. Grouped bar chart with labels <a name="2"></a>

- how to create a grouped bar chart and how to annotate bars with labels

````python
labels = ['G1','G2','G3','G4','G5']
men_means = [20,34,30,35,27]
women_means = [25,32,34,20,25]

x = np.arange(len(labels))
width = 0.35

fig,ax = plt.subplots()
rects1 = ax.bar(x-width/2,men_means,width,label='Men')
rects2 = ax.bar(x+width/2,women_means,width,label='Women')

#add some text for labels, title and custom x-axis tick labels
ax.set_ylabel('Scores')
ax.set_title('Scores by group and gender')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()
plt.show()
````

````python
def autolabel(rects):
    """
    Attach a text label abouve each bar in *rects*, displaying its height
    """
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                   xy=(rect.get_x() + rect.get_width() / 2,height),
                   xytext = (0,3), #3 points vertical offsets
                    textcoords = "offset points",
                    ha = 'center',va='bottom'
                   )
````

````python
labels = ['G1','G2','G3','G4','G5']
men_means = [20,34,30,35,27]
women_means = [25,32,34,20,25]

x = np.arange(len(labels))
width = 0.35

fig,ax = plt.subplots()
rects1 = ax.bar(x-width/2,men_means,width,label='Men')
rects2 = ax.bar(x+width/2,women_means,width,label='Women')

#add some text for labels, title and custom x-axis tick labels
ax.set_ylabel('Scores')
ax.set_title('Scores by group and gender')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()

autolabel(rects1)
autolabel(rects2)

fig.tight_layout()
plt.show()
````

### 3. 补充plt.subplots() <a name="3"></a>

````python
# plt.subplots() returns to fig and ax(s)
````

- Creating multiple subplots using plt.subplots()

````python
x = np.linspace(0,2*np.pi,400)
y = np.sin(x**2)
````

````python
# subplots() without arguments returns a Figure and a single Axes
fig,ax = plt.subplots()
ax.plot(x,y)
ax.set_title('A Single Plot')
plt.show()
````

````python
# define the number of rows or cols in the plots
fig,axs = plt.subplots(2)
fig.suptitle('Vertically Stacked Subplots')
axs[0].plot(x,y)
axs[1].plot(x,-y)
plt.show()
````

````python
# OR
fig, (ax1,ax2) = plt.subplots(2,sharex = True) #two rows, one col and share the same x-axis
fig.suptitle('Vertically Stacked Subplots')
ax1.plot(x,y)
ax2.plot(x,-y)
plt.show()
````

````python
# To obtain side-by-side subplots, pass parameters 1, 2 for one row and two columns
````

````python
fig, (ax1,ax2) = plt.subplots(1, 2,sharey = True) #one row, two columns, share the same y-axis
fig.suptitle('Horizontally Stacked Subplots')
ax1.plot(x,y)
ax2.plot(x,-y)
plt.show()
````

- Stacking subplots in two directions

````python
fig,axs = plt.subplots(2,2)
axs[0,0].plot(x,y)
axs[0,0].set_title('Axis [0,0]')

axs[0,1].plot(x,y,'tab:orange')
axs[0,1].set_title('Axis [0,1]')

axs[1,0].plot(x,-y,'tab:red')
axs[1,0].set_title('Axis [1,0]')

axs[1,1].plot(x,-y,'tab:blue')
axs[1,1].set_title('Axis [1,1]')

for ax in axs.flat:
    ax.set(xlabel = 'x-label',ylabel='y-label')
    
#hide the x labels and tick labels for top plots and y ticks for right plots
for ax in axs.flat:
    ax.label_outer()
````

````python
# OR
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2)
fig.suptitle('Sharing x per column, y per row')
ax1.plot(x, y)
ax2.plot(x, y**2, 'tab:orange')
ax3.plot(x, -y, 'tab:green')
ax4.plot(x, -y**2, 'tab:red')

for ax in fig.get_axes():
    ax.label_outer()
````

````python
fig, (ax1,ax2) = plt.subplots(2) 
fig.suptitle('Axes values are scaled individually by default')
ax1.plot(x,y)
ax2.plot(x + 1,-y)
plt.show()
````

````python
fig, axs = plt.subplots(3, sharex=True, sharey=True)
fig.suptitle('Sharing both axes')
axs[0].plot(x, y ** 2)
axs[1].plot(x, 0.3 * y, 'o')
axs[2].plot(x, y, '+')
plt.show()
````

````python
# Polar axis
fig, (ax1, ax2) = plt.subplots(1, 2, subplot_kw=dict(projection='polar'))
ax1.plot(x, y)
ax2.plot(x, y ** 2)

plt.show()
````

### 5. 设置坐标轴 <a name="5"></a>

````python
x = np.linspace(-1,1,100)
y1 = 2*x + 1 
y2 = x**3

plt.plot(x,y1,c='r',ls = '--',linewidth=1)
plt.plot(x,y2,c='b',ls='-',linewidth=4)
plt.show()
````

````python
x = np.linspace(-3,3,100)
y1 = 2*x + 1 
y2 = x**3

# 改变xy轴的范围
plt.xlim((-1,2))
plt.ylim((-2,3))

plt.plot(x,y1,c='r',ls = '--',linewidth=1)
plt.plot(x,y2,c='b',ls='-',linewidth=4)
plt.show()
````

````python
x = np.linspace(-3,3,100)
y1 = 2*x + 1 
y2 = x**3

# 改变xy轴的范围
plt.xlim((-1,2))
plt.ylim((-2,3))

# 对xy轴的标签进行命名
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')

plt.plot(x,y1,c='r',ls = '--',linewidth=1)
plt.plot(x,y2,c='b',ls='-',linewidth=4)
plt.show()
````
